```python
# Algoritmo de Classificação Machine Learning 

# Neste Script irei desenvolver um algoritmo de classificação para identificar quais as características que influenciam na ocorrência de outros ataques terroristas 

# A base que utilizarei a seguir são apenas de ataques que ocorreram no Iraque de 2000 a 2017
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
uri = '/content/baseTerrorismoMl.csv'
dados = pd.read_csv(uri)
dados.head()
```

```python
# criando dummies

dumAlvo = pd.get_dummies(dados['Descrição do Tipo do Alvo'])
dumArma = pd.get_dummies(dados['Descrição da arma'])
dumAtaque = pd.get_dummies(dados['Descrição do Ataque'])


dumAlvo.head()
```

```python
dumArma.head()
```

```python
dumAtaque.head()
```

```python
dados01 = pd.concat([dados,dumAlvo,dumArma,dumAtaque],axis=1) # dentro do colchete vc passa uma lista dos datasets, o axis é se vc quer juntar as tabelas na horizontal (1) ou na vertical (0)
dados01.head()
```

```python
# Selecionando as colunas que eu quero trabalhar
x = dados01 [["Ataque suicida?","Ataque a Propriedade", "Houve refém?","Airports & Aircraft","Business","Food or Water Supply","Government (General)","Military","Police","Private Citizens & Property","Terrorists/Non-State Militia","Utilities","Violent Political Party","Chemical","Explosives","Firearms","Incendiary","Melee","Assassination","Bombing/Explosion","Facility/Infrastructure Attack","Hijacking","Hostage Taking (Kidnapping)"]]
y = dados01 [["multiple"]]
```

```python
# Quando vamos treinar e testar o algoritmo, precisamos separar os dados de treino e de teste
# Primeiro vamos analisar a quantidade de dados
dados01.shape
```

```python
#Vamos separar 75% para treinar e o resto será para testar
treino_x = x[:1415]
treino_y = y[:1415]
#Agora vamos pegar o resto para teste
teste_x = x[1415:]
teste_y = y[1415:]
```

```python
# Primeiro vamos importar o módulo e depois treinar o nosso modelo
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)

#Agora vamos definir o teste
previsoes = modelo.predict(teste_x)

#Agora vamos calcular a acuracidade
acuracia = accuracy_score(teste_y, previsoes) *100
print("A acurácia foi %.2f" % acuracia)
```
