```python
#Importar as bibliotecas para tratar os dados
import pandas as pd
import seaborn as srn
import statistics as sts
```

```python
#Importar os dados para realizar os tratamentos
#A Base de Dados sempre será disponibilizada e atualizada no mesmo diretório
dsFato = pd.read_csv("fato.csv", sep=";")
dsFato.head()
```

```python
# Existem algumas colunas que não serão utilizadas no dashboard, por isso serão removidas
dsFato = dsFato.drop(columns=['nkillus','nwoundus','nwoundte'])
dsFato.head()
```

```python
#Agora alguns cabeçalhos que estão em inglês, serão traduzidos
dsFato.rename(columns={'suicide': 'Ataque suicida?'}, inplace = True)
dsFato.rename(columns={'corp1': 'Alvo'}, inplace = True)
dsFato.rename(columns={'nkillter': 'Terroristas mortos'}, inplace = True)
dsFato.rename(columns={'nwound': 'Número de feridos'}, inplace = True)
dsFato.rename(columns={'hostkidoutcome_txt': 'Resumo do resgate'}, inplace = True)
dsFato.head()
```

```python
#Agora será identificado os dados nulos
dsFato.isna().mean().round(4)
```

```python
#Preencher os valores nulos com 'Unknown' para manter um padrão
dsFato.update(dsFato.fillna('Unknown'))
dsFato.isna().mean().round(4)
```

```python
#A partir deste ponto será realizada uma análise exploratória dos dados para que seja identificado os possíveis pontos que deverão ser tratados
```

```python
#tamanho da base
dsFato.shape
```

```python
agroup = dsFato.groupby(['Ataque suicida?']).size()
agroup
```

```python
a = {'Unknown':'não', 0:'não', 1:'sim'}
dsFato['Ataque suicida?'] = dsFato['Ataque suicida?'].map(a)
```

```python
Agrupado = dsFato.groupby(['Ataque a Propriedade']).size()
Agrupado
```

```python
a = {'Unknown':'não', 0:'não', 1:'sim','-9':'não'}
dsFato['Ataque a Propriedade'] = dsFato['Ataque a Propriedade'].map(a)
```

```python
#As colunas ID devem ser número inteiros, por isso o que for diferente de um número intero será substituído por zero
dsFato['ID_DescricaoPropriedade'] = dsFato['ID_DescricaoPropriedade'].replace('Unknown','0')
dsFato['ID_GrupoTerror'] = dsFato['ID_GrupoTerror'].replace('Unknown','0')
dsFato['ID_TipoCrime'] = dsFato['ID_TipoCrime'].replace('Unknown','0')
dsFato['ID_Pais'] = dsFato['ID_Pais'].replace('Unknown','0')
dsFato['ID_Regiao'] = dsFato['ID_Regiao'].replace('Unknown','0')
dsFato['ID_TipoAtaque'] = dsFato['ID_TipoAtaque'].replace('Unknown','0')
dsFato['ID_TipoAlvo'] = dsFato['ID_TipoAlvo'].replace('Unknown','0')
dsFato['ID_SubTipoAlvo'] = dsFato['ID_SubTipoAlvo'].replace('Unknown','0')
dsFato['ID_TipoArma'] = dsFato['ID_TipoArma'].replace('Unknown','0')
dsFato['ID_SubTipoArma'] = dsFato['ID_SubTipoArma'].replace('Unknown','0')

dsFato['ID_DescricaoPropriedade'] = dsFato['ID_DescricaoPropriedade'].replace('#N/D','0')
dsFato['ID_GrupoTerror'] = dsFato['ID_GrupoTerror'].replace('#N/D','0')
dsFato['ID_TipoCrime'] = dsFato['ID_TipoCrime'].replace('#N/D','0')
dsFato['ID_Pais'] = dsFato['ID_Pais'].replace('#N/D','0')
dsFato['ID_Regiao'] = dsFato['ID_Regiao'].replace('#N/D','0')
dsFato['ID_TipoAtaque'] = dsFato['ID_TipoAtaque'].replace('#N/D','0')
dsFato['ID_TipoAlvo'] = dsFato['ID_TipoAlvo'].replace('#N/D','0')
dsFato['ID_SubTipoAlvo'] = dsFato['ID_SubTipoAlvo'].replace('#N/D','0')
dsFato['ID_TipoArma'] = dsFato['ID_TipoArma'].replace('#N/D','0')
dsFato['ID_SubTipoArma'] = dsFato['ID_SubTipoArma'].replace('#N/D','0')

dsFato['ID_DescricaoPropriedade']=dsFato['ID_DescricaoPropriedade'].astype(int)
dsFato['ID_GrupoTerror']=dsFato['ID_GrupoTerror'].astype(int)
dsFato['ID_TipoCrime']=dsFato['ID_TipoCrime'].astype(int)
dsFato['ID_Pais']=dsFato['ID_Pais'].astype(int)
dsFato['ID_Regiao']=dsFato['ID_Regiao'].astype(int)
dsFato['ID_TipoAtaque']=dsFato['ID_TipoAtaque'].astype(int)
dsFato['ID_TipoAlvo']=dsFato['ID_TipoAlvo'].astype(int)
dsFato['ID_SubTipoAlvo']=dsFato['ID_SubTipoAlvo'].astype(int)
dsFato['ID_TipoArma']=dsFato['ID_TipoArma'].astype(int)
dsFato['ID_SubTipoArma']=dsFato['ID_SubTipoArma'].astype(int)
```

```python
Agrupado = dsFato.groupby(['Houve refém?']).size()
Agrupado
```

```python
a = {'Unknown':'não', 0:'não', 1:'sim','-9':'não'}
dsFato['Houve refém?'] = dsFato['Houve refém?'].map(a)
```

```python
#Agora as informações das dimensões serão inseridas na fato de acordo com os respectivos ID's
```

```python
df3= pd.read_csv('dimensaoRegiao.csv', sep=";")
dw = pd.merge(dsFato, df3, on = 'ID_Regiao', how = 'inner')
```

```python
df4= pd.read_csv('dimensaoSubTipoDeAlvo.csv', sep=";")
dw1 = pd.merge(dw, df4, on = 'ID_SubTipoAlvo', how = 'inner')
```

```python
df5= pd.read_csv('dimensaoSubTipoDeArma.csv', sep=";")
dw2 = pd.merge(dw1, df5, on = 'ID_SubTipoArma', how = 'inner')
```

```python
df6= pd.read_csv('dimensaoTipoDeAlvo.csv', sep=";")
dw3 = pd.merge(dw2, df6, on = 'ID_TipoAlvo', how = 'inner')
```

```python
df7= pd.read_csv('dimensaoTipoDeArma.csv', sep=";")
dw4 = pd.merge(dw3, df7, on = 'ID_TipoArma', how = 'inner')
```

```python
df8= pd.read_csv('dimensaoTipoDeAtaque.csv', sep=";")
dw5 = pd.merge(dw4, df8, on = 'ID_TipoAtaque', how = 'inner')
```

```python
df9= pd.read_csv('dimensaoTipoDeCrime.csv', sep=";")
dw6 = pd.merge(dw5, df9, on = 'ID_TipoCrime', how = 'inner')
```
